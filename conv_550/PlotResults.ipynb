```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import matplotlib
import numpy as np
import matplotlib.pyplot as plt

#data = numpy.loadtxt(open("results.csv", "rb"), 
#                   dtype={'names': ('Algorithm', 'ImageSize', 'KernelLength', 'Throughput', 'Time'), 'formats': ('S20', 'i4', 'i4', 'f4', 'f4')}, delimiter=';')

data = np.loadtxt(open("resultsThroughput.csv", "rb"), 
                   dtype={'names': ('Kernel','CPU','NVIDIA','FullNaive','SepNaive','SepShared','SharedUnroll','SharedTile','SharedTileC'),
                        'formats': ('i4', 'f4', 'f4', 'f4', 'f4', 'f4', 'f4', 'f4', 'f4')}, delimiter=';', skiprows=1)
data.shape
```

```python
plt.rcParams["figure.figsize"] = [12,5]
```

```python

# data to plot
n_groups = data.shape[0]
 
# create plot
fig, ax = plt.subplots()
index = np.arange(n_groups)
bar_width = 0.1
 
plt.bar(index, list(data['CPU']), bar_width, label='CPU')
plt.bar(index+1*bar_width, list(data['FullNaive']), bar_width, label='1: full naive')
plt.bar(index+2*bar_width, list(data['SepNaive']), bar_width, label='2: seperable naive')
plt.bar(index+3*bar_width, list(data['SepShared']), bar_width, label='3: seperable shared')
plt.bar(index+4*bar_width, list(data['SharedUnroll']), bar_width, label='4: shared unroll')
plt.bar(index+5*bar_width, list(data['SharedTile']), bar_width, label='5: shared tiling')
plt.bar(index+6*bar_width, list(data['SharedTileC']), bar_width, label='6: tiling coalesced')
plt.bar(index+7*bar_width, list(data['NVIDIA']), bar_width, label='NVIDIA')
 
plt.xlabel('Kernel Length, Pixels', fontsize=18)
plt.ylabel('Throughput, MPixels/sec', fontsize=18)
plt.title('Throughput by Kernel Length', fontsize=20)
plt.xticks(index + bar_width, data['Kernel'])
plt.legend(bbox_to_anchor=(1.01, 1), loc=2, borderaxespad=0.)
 
plt.tight_layout()
plt.show()
```
